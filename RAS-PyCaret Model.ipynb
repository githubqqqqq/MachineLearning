 All the prepprocessing steps are applied within setup(). PyCaret creates a transformation pipeline based on the parameters defined in setup function. It automatically orchestrates all dependencies in a pippeline so that there is no need to manually manage the sequential execution of transformations on test dataset. PyCaret's pipeline can easily be transferred across environments to run at scale or be deployed in production with ease. Following are transformations within the first release:
Sample and Split
Missing Value Imputation
One-Hot Encoding
Normalization
Transformation
Interaction, Binning
Feature Importance
Create Clusters
Remove outliers

```python
def create_models(list_of_models):
    for model in list_of_models:
        i=list_of_models.index(model)+1
        print_result="-----creating model "+model+"-----"
        print(model)
        df=pd.DataFrame([model])
        model=create_model(model)
        ##plot_model(model, plot = 'auc')
        sheet_name='M'+str(i)
        worksheet = sh.worksheet_by_title(sheet_name)
        worksheet.clear("A1","M5000")
        worksheet.set_dataframe(df, "A1")
        try: 
            feature_importance=model.feature_importances_.tolist()     
            feature_importance_list=list(zip(features,feature_importance))
            feature_importance_df = pd.DataFrame (feature_importance_list, columns = ['Feature','Importance'])
            feature_importance_df=feature_importance_df.sort_values(by='Importance',ascending=False)

            worksheet.set_dataframe(feature_importance_df, "B1")
        except:
            print ("-----this model has not feature importance-----")
        pred=predict_model(model,df_sub,raw_score=True)
        pred['Decile'] = pd.qcut(pred['Score_1'], 10,labels = False,duplicates='drop')+1
        decile1=pred[["Decile","Y"]].groupby("Decile",dropna=False).count()
        decile2=pred[["Decile","Y"]].groupby("Decile",dropna=False).sum()
        decile_all=decile1.merge(decile2,on='Decile',how='left')
        decile_all['pct_1']=decile_all.Y_y/decile_all.Y_x
        decile_all=decile_all.sort_values(by='Decile',ascending=False).reset_index()
        worksheet.set_dataframe(decile_all, "E1")
                    
            
        scores_1=pred[['Score_1','Y']].groupby("Score_1").count()
        scores_1=scores_1.sort_values(by="Y",ascending=False).reset_index()
        scores_1_100=scores_1.head(100)
        
        worksheet.set_dataframe(scores_1_100, "I1")
```

# Ensemble Model

The ensemble_model function is used for ensembling trained models. It takes only one parameter. ie. a trained model object. This function returns a table with k-fold cross validated scores and a trained model object. Bagging method is used for ensembling by default which can be changed to 'Boosting'' by using the method parameter within the ensemble_model function

Bagging involves fitting many decision trees on different samples of the same dataset and averaging the predictions.
Stacking involves fitting many different models types on the same data and using another model to learn how to best combine the predictions.
Boosting involves adding ensemble members sequentially that correct the predictions made by prior models and outputs a weighted average of the predictions.

https://gist.github.com/pycaret/e21f838a26923856db896df10cc3c7e6
